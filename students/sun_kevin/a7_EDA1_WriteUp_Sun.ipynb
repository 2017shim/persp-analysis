```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_csv("gss2012.csv")
```

## Introduction

The General Social Survey provides a wealth of not only demographic information, but information highlighting Americans’ viewpoints on a wide spectrum of social, economic, cultural, and general policy issues. From conducting the exploratory data analysis on this dataset, we can glean significant insight and make countless comparisons across varying sub-sections of the American population. Collected in 2012, this dataset captures sentiment of Americans at a novel time in the history of the country – in the midst of the first non-white President’s administration. While sentiment towards race relations and cohesion in the United States generally improved over the latter half of the 20th-century, we can utilize the data from the General Social Survey from 2012 to potentially answer research questions that address deeper issues surrounding race. 

## Potential Research Questions

Some key questions that could be addressed by researchers through this particular data set include:
-	Does political party affiliation impact different racial groups views on race-conscious policies (i.e. affirmative action)?
-	To what extent are Americans tolerant of racism, racial prejudice, or racial discrimination; and in what arenas are Americans tolerant of such sentiments?
-	To what extent might interpersonal relationships between white and black Americans reflect the broader populations views on racism, race-conscious policies, etc.?

## Contextualizing the Data Set and Justifying Research Questions

From conducting the exploratory data analysis (EDA), one can observe potential sampling biases that may have an impact on responses and thus, overall patterns in opinion. For example, the EDA revealed an oversampling of white Americans and under-sampling of Americans of color. Especially when considering policies and sentiment relating directly to race relations and race-conscious policies, this bias holds significance. To aggregate every racial group that is not white or black into “other” erases the potential to compare and assess more nuanced racial sentiment. 

While a more detailed and nuanced picture is unavailable through this particular data set, as researchers, we are still able to explore research questions that are quite specific in their scope. Given the availability of peoples’ reported race, political affiliation, views on achievement/hard work, affirmative action, and relationship/closeness to blacks/whites we are able to gain insight into the above research questions through multi-panel and cross-demographic visualizations. 



## Political Party Affiliation and Race-Conscious Policies 

Observing the swarm plot below, we are able to confirm well-established knowledge of political party affiliation compared by race. We see the largest proportion of black voters as StrDem (strong Democrats), WkDem (weak Democrats), or Ind (Independent), with a smaller proportion as IndDem (Independent Democrats). While other Americans of color identify most with being WkDem and Ind. We can further observe that StrRep (strong Republican) and WkRep (weak Republican) are dominated by white voters. 


```python
sns.swarmplot(x="partyid", y="age", hue="race", data=df, dodge=True, size=3.5, palette='cubehelix'); 
```

We can further zoom in on public sentiment surrounding affirmative action as a race-conscious policy while simultaneously comparing support/opposition in regard to race and political party affiliation. In the multi-panel swarm plot below, we can differentiate support levels towards affirmative action (y-axis) by each political party affiliation (individual graphs). An immediate insight one learns is that regardless of political party affiliation, the majority of Americans in each identified party, either report opposition or strong opposition to affirmative action as a policy. Furthermore, while it may not be a surprise that white Americans are by-and-large in opposition to affirmative action, we can observe that both Black and Other-identified Americans are not overwhelmingly in favor of affirmative action either. In fact, even for independent Democrats, weak Democrats, and independents who are non-white, a slightly majority oppose affirmative action. Perhaps most surprisingly, even among non-white strong Democrats, sentiment is evenly divided over support for affirmative action. 

Therefore, even strong partisanship in policy such as affirmative action (which we view as policy more strongly supported by Democrats than Republicans) is not reflected in Americans sentiment. This would then inform further investigation into other overriding factors that shape Americans views on race and race-related policies, leading to our subsequent research questions.


```python
sns.factorplot(x="age", y="affrmact", hue="race", col = "partyid", data=df, kind="swarm", size=3, dodge=True, palette='cubehelix');
```

## The Space and Extent of Public Tolerance Towards Racism & Racial Prejudice

So we may further hypothesize that views towards race-conscious policies are shaped over longer periods of time through peoples education, interactions, and the norms that come to be established through society. Generally, we can assume that when asked to assess a statement similar to, “Racism and racial prejudice has no place in American society,” that the vast majority of Americans would agree with the statement. While we could delve deeper into social desirability bias of such an expected response, we will not do so in this space. Instead, we can directly observe Americans’ tolerance for racism in a space that is vital to shaping such racial views – the classroom. 

The graph below demonstrates significant levels of tolerance the general public holds for allowing a person with racist beliefs to teach at a university of college. Although majorities of black and other non-white Americans are in favor of not allowing such a person to teach at a university, a slight majority of white Americans are in favor. Despite most Americans disapproval of racism, overriding factors such as beliefs in freedom of speech and open discourse, regardless of the extremity, intolerance, and bigotry appear to take priority. 


```python
sns.countplot(x="colrac", hue="race", data=df, palette="Set2");
```

In further considering whether racist speech in communities and distribution of racist literature in libraries should be allowed, we continue to observe higher tolerance rates from white Americans, but also significant levels of tolerance and acceptance from non-white Americans as well. Thus, we learn that while racism and prejudice is broadly unacceptable to most Americans, in practice, such beliefs are not disqualifying factors. 

```python
sns.countplot(x="spkrac", hue="race", data=df, palette="Set2");
```

```python
sns.countplot(x="librac", hue="race", data=df, palette="Set2");
```

## Interpersonal Relationships

If we examine interpersonal relationships between whites, blacks, and other non-whites this may inform such prevalence of non-disqualifying racist beliefs.  Strong arguments can be made that peoples’ viewpoints on issues are mostly informed through the interactions and relationships they have with those in the immediate circle. To better inform our findings on support for race-conscious policies and tolerance of racism in arenas of speech, education, or publically-funded literature, examining interpersonal relationships and sense of “closeness” between races may hold great utility. 

As seen in the below graphs, when surveyed on perceptions of how close one feels to whites and blacks, there is significant disparity in white closeness to other whites versus blacks. While similar amounts of whites reported ambivalent (around 5.0) feelings towards white and blacks, the proportion of whites reporting closeness to other whites is 3 times greater than whites reporting closeness to blacks. Similarly, blacks report much higher levels of closeness to other blacks than whites. 


```python
sns.countplot(x="closewht", hue="race", data=df);
```

```python
sns.countplot(x="closeblk", hue="race", data=df);
```

## Conclusions

Overall, from conducting this EDA, we were able to find that feelings of closeness between whites and blacks are lacking relative feelings of closeness among the same racial group. One may hypothesize that such interpersonal sentiment would translate to broader patterns of support or opposition to policies that are explicitly based on race. This was observed through an analysis of support on affirmative action policies. Despite such policy being highly partisan in nature, the reported supporting and opposition rates did not appear to differ significantly when we split the graphs into political party affiliation. While blacks are much more likely to support affirmative action, they do not support in overwhelming numbers. We might further conclude that such sentiment is buttressed by a relatively high level of tolerance for racist expression and tolerance for people who harbor racist beliefs to work in environments, like the classroom, that shape the general populaces’ beliefs on race issues.
