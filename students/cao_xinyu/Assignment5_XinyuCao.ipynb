### Problem 3

The is the dataset of median income of US Cities. In the following, I will draw a plot of the fitted the distribution of median income of over the more than 25,000 cities in US. We could see a large inequality across USA cities. There are cities very poor has lower than 20,000 of median household income, and with extremly wealthy cities, that can be barely seen from the vilion plot.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
from scipy import stats

income = pd.read_csv('MedianHouseholdIncome2015.csv', encoding="windows-1252")
xincomes = list(income['Median Income'].dropna())
yincomes = []
for xincome in xincomes:
    try:
        yincomes.append(int(xincome))
    except ValueError:
        pass
sns.distplot(yincomes, bins=30)
plt.title('median income distribution of US Cities')
plt.xlabel('Median Income of Cities($)')
plt.ylabel('Density')
plt.show()
```

```python
sns.violinplot(yincomes, bins=30)
plt.title('median income distribution of US Cities')
plt.xlabel('Median Income of Cities($)')
plt.ylabel('Density')
plt.show()
```
