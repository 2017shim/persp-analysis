# Data Preperation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
raw = pd.read_csv('gss2012.csv')
mask = raw['income06'].notnull()
df = raw[mask]
income_dict = {'$150000 OR OVER':15, '$110000 TO $129999':10, '$130000 TO $149999':13,
       '$50000 TO 59999':5, '$25000 TO 29999':2.5, '$30000 TO 34999':3.0,
       '$5 000 TO 5 999':0.5, 'UNDER $1 000':0.0, '$8 000 TO 9 999':0.8,
       '$22500 TO 24999':2.25, '$40000 TO 49999':4, '$60000 TO 74999':6,
       '$90000 TO $109999':9, '$75000 TO $89999':7.5, '$35000 TO 39999':3.5,
       '$15000 TO 17499':1.5, '$20000 TO 22499':2, '$6 000 TO 6 999':0.6,
       '$17500 TO 19999':1.75, '$12500 TO 14999':1.25, '$10000 TO 12499':1,
       '$1 000 TO 2 999':0.1, '$7 000 TO 7 999':0.7, '$4 000 TO 4 999':0.4,
       '$3 000 TO 3 999':0.3}
df['income'] = df['income06'].apply(lambda x: income_dict[x])
```

# Introduction 
The General Social Survey provides a survey data on American's basic demographic information, such as income, sex, race and party affliation. We want to understand if there is a inequality of different group of people. 

```python
sns.countplot(x='income', data=df)
plt.xlabel('Density')
plt.ylabel('Individual Income (in thousands of dollar)')
plt.title('Fig1: Income Distribution of Whole Data')
plt.show()
```

```python
sns.countplot(x='income', hue='sex', data=df)
plt.xlabel('Individual Income (in thousands of dollar)')
plt.ylabel('Count')
plt.title('Fig2: Income Distribution by Gender')
plt.show()
```

```python
sns.countplot(x='income', hue='voted08', data=df)
plt.ylabel('Count')
plt.xlabel('Individual Income (in thousands of dollar)')
plt.title('Fig3: Income Distribution by 2008 Voting')
plt.show()
```

```python
sns.boxplot(x='partyid_3', y='income', data=df)
plt.xlabel('Party Affliation')
plt.ylabel('Individual Income (in thousands of dollar)')
plt.title('Fig4: Income Distribution by Different Party Affliation')
plt.show()
```

```python
sns.violinplot(x='race', y='income', data=df)
plt.xlabel('Race')
plt.ylabel('Individual Income (in thousands of dollar)')
plt.title('Fig5: Income Distribution by Race')
plt.show()
```

```python
sns.boxplot(x='degree', y='income', data=df)
plt.xlabel('Highest Degree')
plt.ylabel('Individual Income (in thousands of dollar)')
plt.title('Fig6: Income Distribution by Education Level')
plt.show()
```
